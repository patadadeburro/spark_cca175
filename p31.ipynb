#### Answer

create local directory

```
mkdir /tmp/nrtcontent
```

create flume configuration file
`p31.conf`

```
a1.sources                   = r1
a1.sinks                     = k1
a1.channels                  = c1

# sources
a1.sources.r1.type           = spooldir
a1.sources.r1.spoolDir       = /tmp/nrtcontent

# sinks
a1.sinks.k1.type             = hdfs
a1.sinks.k1.hdfs.path        = /tmp/flume
a1.sinks.k1.hdfs.fileType    = DataStream
a1.sinks.k1.hdfs.filePrefix  = events
a1.sinks.k1.hdfs.fileSuffix  = .log
a1.sinks.k1.hdfs.inUsePrefix = _

# channels
a1.sources.r1.channels       = c1
a1.sinks.k1.channel          = c1
a1.channels.c1.type          = memory
```

start Flume agent

```
flume-ng    agent
--conf      /home/cloudera/flumeconf 
--conf-file /home/cloudera/flumeconf/p31.conf 
--name      a1
```

open another linux terminal and create txt files

```
cd /tmp/nrtcontent
echo "I am preparing for CCA175 from ABCTECH.com" > /tmp/nrtcontent/.he1.txt 
mv /tmp/nrtcontent/.he1.txt /tmp/nrtcontent/he1.txt
```

After few mins

```
echo "I am preparing for CCA175 from TopTech.com" > /tmp/nrtcontent/.qt1.txt 
mv /tmp/nrtcontent/.qt1.txt /tmp/nrtcontent/qt1.txt
```
