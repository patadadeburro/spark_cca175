#### Problem Scenario 90 : 
You have been given below two files

course.txt

|||
|-|-|
|id | course |

```
1,Hadoop
2,Spark
3,HBase
```

fee.txt

|||
|-|-|
|id | fee |

```
2,3900
3,4200
4,2900
```

Accomplish the following activities.
1. Select all the courses and their fees , whether fee is listed or not.
2. Select all the available fees and respective course. If course does not exists still list the fee
3. Select all the courses and their fees , whether fee is listed or not. However, ignore records having fee as null.

#### Answer

```
import org.apache.spark.SparkContext
import org.apache.spark.SparkContext._
import org.apache.spark.SparkConf

import org.apache.spark.sql
import org.apache.spark.sql._
import org.apache.spark.sql.functions

val sqlContext = new SQLContext( sc )

import sqlContext.implicits
import sqlContext.implicits._
```

create case classes

```
case class Course( id: Int, name: String )
case class Fee   ( id: Int, fee : Int    )
```

create rdd of objects

```
val c_rdd = sc.textFile( "p32/course.txt" ).map( _.split( "," ) ).map( a => Course( a(0).toInt, a(1) ) )
val f_rdd = sc.textFile( "p32/fee.txt"    ).map( _.split( "," ) ).map( a => Fee   ( a(0).toInt, a(1).toInt ) )
```

create dataframes

```
val course = c_rdd.toDF()
val fee    = f_rdd.toDF()
```

register tables to make queries

```
course.registerTempTable( "course" )
fee.registerTempTable   ( "fee"    )
```

1 . Select all the courses and their fees , whether fee is listed or not.

```
val r1 = sqlContext(
"""
SELECT c.id, c.name, f.fee 
FROM   course as c LEFT JOIN fee as f ON c.id = f.id
""" )
r1.show()
```

2 . Select all the available fees and respective course. If course does not exists still list the fee

```
val r2 = sqlContext.sql(
"""
SELECT c.name, f.id, f.fee 
FROM   course as c RIGHT JOIN fee as f ON c.id = f.id
""" )
r2.show()
```

3 . Select all the courses and their fees , whether fee is listed or not. However, ignore records having fee as null.

```
val r3 = sqlContext.sql(
"""
SELECT c.id, c.name
FROM   course as c LEFT JOIN fee as f ON c.id = f.id
WHERE  f.id IS NOT NULL
""" )
r3.show()
```
