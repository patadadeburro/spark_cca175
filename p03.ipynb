#### Step 1 : 
Select all the products which has product code as null

```
# work around for clouder quick start virtual machine.
import org.apache.spark.sql._
import org.apache.spark.sql.SQLContext
import org.apache.spark.sql.hive._
import org.apache.spark.sql.hive.HiveContext


val sqlContext = new HiveContext(sc)
import sqlContext.implicits._
```

```
// spark-shell
val results = sqlContext.sql( "SELECT * FROM products WHERE code IS NULL" ) 
results.show()

val results = sqlContext.sql( "SELECT * FROM products WHERE code IS NULL ",,M ) 
results.show()
```

```
# pyspark
results = sqlContext.sql( "SELECT * FROM products WHERE code IS NULL" ) 
results.show()
```

## Step 2 : 
Select all the products , whose name starts with Pen and results should be order by Price descending order.

```
// spark-shell
val results = sqlContext.sql( "SELECT * FROM products WHERE name LIKE 'Pen %' ORDER BY price DESC " ) 
results. show()
```

```
# pyspark
results = sqlContext.sql( "SELECT * FROM products WHERE name like 'PEN' ORDER BY price DESC" ) 
results.show()
```

## Step 3 :

```
// spark-shell
val results = sqlContext.sql( "SELECT * FROM products WHERE name LIKE 'Pen %' ORDER BY price DESC, quantity" ) 
results.show()                            
```

```
# pyspark
results = sqlContext.sql( "SELECT * FROM products WHERE name LIKE 'Pen %' ORDER BY price DESC, quantity" ) 
results.show()
```

## Step 4 :
