#### Activity 1
It is possible that, same product can be supplied by multiple supplier. Now find each product, its price according to each supplier.

```
r1 = sqlContext.sql( """ 
SELECT ps.productID, ps.supplierID
       p.name as product_name, price, 
       s.name as supplier_name
FROM products_suppliers as ps
JOIN product  as p ON ps.productID  = p.productID
JOIN supplier as s ON ps.supplierID = s.supplierID
ORDER BY p.productID
""" )
r1.show()
```

#### Activity 2
Find all the supllier name, who are supplying 'Pencil 3B'

```
r2 = sqlContext.sql( """ 
SELECT ps.productID, ps.supplierID
       p.name as product_name,
       s.name as supplier_name
FROM products_suppliers as ps
JOIN product  as p ON ps.productID  = p.productID
JOIN supplier as s ON ps.supplierID = s.supplierID
WHERE p.name = 'Pencil 3B'
""" )
r2.show()
```
